```python
import yfinance as yf

def get_clean_market_data():
    # 1. Define Tickers
    tickers = ["^NSEI", "^INDIAVIX"] # for nifty and volatility 
    print("Fetching 10 years of data...")
    raw_data = yf.download(tickers, start="2016-01-01", end="2026-01-15")

    # 2. Extract Nifty 50 (OHLCV)
    nifty = raw_data.xs('^NSEI', axis=1, level=1)
    
    # 3. Extract India VIX 
    vix = raw_data.xs('^INDIAVIX', axis=1, level=1)[['Close']]
    vix.columns = ['VIX_Close']

    # 4. Combine them into one Organized Table
    combined_data = nifty.join(vix)

    # 5. Cleaning: Remove 'Missing' data
    clean_data = combined_data.dropna()
    
    # 6. Final Polish
    clean_data.reset_index(inplace=True)

    return clean_data

final_data = get_clean_market_data()

print("\n--- Cleaned Data Preview (OHLCV + VIX) ---")
print(final_data.head(500))

final_data.to_csv("cleaned_nifty_vix_10y.csv", index=False)
print("\nFile 'cleaned_nifty_vix_10y.csv' is ready!") #Saving as .csv for future purposes 
```
